# Matematiikkaa Pythonilla, numpy-kirjasto

Python on varsin tehokas työkalu matemaattisten ongelmien ratkaisemisessa. Sen numpy-kirjasto sisältää siihen liittyen hyödyllisiä työkaluja. Jos käyttämäsi ohjelma ei tunnista kirjastoa, se täytyy asentaa. Tämä vaihtelee tietysti ohjelmakohtaisesti, esimerkiksi PyCharmissa kun käytät uutta kirjastoa, se ilmoittaa, että tämä pitäisi asentaa ja kun kuittaat asennuksen, niin se on sitten siinä.

Vaikka tässä numpyä käytetäänkin, on syytä myös mainita `math`-kirjasto (https://docs.python.org/3/library/math.html), joka sisältää hyödyllisiä matemaattisia funktioita.

Lähdetään kulmien muuttamisesta asteiksi taikka radiaaneiksi. Esimerkiksi radiaaneina esitetyt kulmat

α = 1.35, β = 2.765

saa asteiksi näin: (suorita koodi painamalla kolmiota vasemmalla)

```python
import numpy as np

a = 1.35
b = 2.765

print(np.degrees(a))
print(np.degrees(b))
```

Tämän perusteella on varmaankin kohtuullisen helppo tehtävä ratkaista, miten vaikkpa seuraavat kulmat

α = 48° ja β = 171°

muunnetaan radiaaneiksi. Joka tapauksessa seuraava linkki auttaa asiaan:

https://numpy.org/doc/stable/reference/routines.math.html#trigonometric-functions

# Numpy arrays

Numpyn yksi hyödyllinen ominaisuus on taulukkorakenne. Oikeastaan sen ominaisuudet tulevat esille paremmin vasta sitten kun päästään matriisilaskentaan. Mutta voi sitä käyttää näissäkin esimerkeissä

```python
import numpy as np

A = np.array([0, 90, 180])

for i in A:
  print(np.radians(i))
```
